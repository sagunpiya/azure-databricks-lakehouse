```python
# Read ADLS:
# Note: Serverless compute requires credentials configured via Unity Catalog external locations
# or workspace-level storage credentials, not spark.conf.set()
bronzedf = spark.read.parquet(
"/Volumes/dev/default/files/orderstech"
)
from pyspark.sql.functions import *
bronzedf = bronzedf \
.withColumn(
"LoadDate",
current_timestamp()
)

bronzedf.write \
.format("delta") \
.mode("append") \
.saveAsTable(
"dev_sales.bronze.orders"
)
display(bronzedf)
bronzedf.count()
```
